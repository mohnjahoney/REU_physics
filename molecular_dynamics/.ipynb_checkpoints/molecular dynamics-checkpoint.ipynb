```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

# Simple Harmonic Oscillator - simple solver

```python
def SHO_euler_step(x, v, k, m, dt):
    # Present as playing with the order of the computation of x and v as a way toward leapfrog.
    net_force = -k * x
    a = net_force / m
    x += v * dt
    v += a * dt
    return x, v
```

```python
# Choose initial conditions
x = 3.0
v = 0.0
k = 2.0
m = 7.0
```

```python
# Run two experiments and compare.

fig = plt.figure(figsize=(16, 4))

# Experiment 1
N = 1000
dt = 0.05

xs = [x]
vs = [v]

for step in range(N-1):
    newx, newv = SHO_euler_step(x=xs[step], v=vs[step], k=k, m=m, dt=dt)
    xs.append(newx)
    vs.append(newv)

times = dt*np.arange(N)
plt.plot(times, xs, '.-', label='dt = {}'.format(dt))

N = 5000
dt = 0.01

# Experiment 2

xs = [x]
vs = [v]

for step in range(N-1):
    newx, newv = SHO_euler_step(x=xs[step], v=vs[step], k=k, m=m, dt=dt)
    xs.append(newx)
    vs.append(newv)

times = dt*np.arange(N)
plt.plot(times, xs, '.-', label='dt = {}'.format(dt))

plt.legend()
```

```python
def SHO_energy(x, v, k, m):
    return 0.5 * m * v**2 + 0.5 * k * x**2
```

```python
# Run two experiments and compare.
# TRY: Here, energies are computed with a list comprehension. Can you compute in a different / cleaner / faster way?

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(16, 8), sharex=True)

# Experiment 1
N = 1000
dt = 0.05

xs = [x]
vs = [v]

for step in range(N-1):
    newx, newv = SHO_euler_step(x=xs[step], v=vs[step], k=k, m=m, dt=dt)
    xs.append(newx)
    vs.append(newv)

times = dt*np.arange(N)
energies = [SHO_energy(xs[ind], vs[ind], k, m) for ind in range(N)]
ax[0].plot(times, xs, '.-', label='dt = {}'.format(dt))
ax[1].plot(times, energies, '.-', label='dt = {}'.format(dt))

N = 5000
dt = 0.01

# Experiment 2

xs = [x]
vs = [v]

for step in range(N-1):
    newx, newv = SHO_euler_step(x=xs[step], v=vs[step], k=k, m=m, dt=dt)
    xs.append(newx)
    vs.append(newv)

times = dt*np.arange(N)
energies = [SHO_energy(xs[ind], vs[ind], k, m) for ind in range(N)]
ax[0].plot(times, xs, '.-', label='dt = {}'.format(dt))
ax[1].plot(times, energies, '.-', label='dt = {}'.format(dt))

# TODO
# Compare with analytic solution.
# real_xs = []
# ax[0].plot
# ax[1].plot

ax[1].set_ylim(bottom=0)
ax[0].legend()
ax[1].legend()
```
