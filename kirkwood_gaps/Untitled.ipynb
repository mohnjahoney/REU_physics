```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from scipy import integrate
```

```python
def f(t, pt):
    """
    return derivative
    d x / dt = v_x
    d y / dt = v_y
    d v_x / dt = a_x
    d v_y / dt = a_y
    
    let F = central gravitational force
    Note: force is velocity independent
    """
    
    x, y, v_x, v_y = pt
    
    r = np.sqrt(x**2 + y**2)
    
    m_central = 1.5
    
    # F/m
    G = 1.0
    
    a_mag = G * m_central / r**2
    
    a_x = - a_mag * x / r
    a_y = - a_mag * y / r
    
    return pt[2], pt[3], a_x, a_y # needs to be nparray?
```

```python
t0 = 0
T = 10
Nt = 500
pt0 = [3, 0, 0, 0.5]
ts = np.linspace(t0, T, Nt)

sol = integrate.solve_ivp(fun=f, t_span=[t0, T], y0=pt0, t_eval=ts, method='RK23')
```

```python
fig = plt.figure(figsize=(12, 12))
ax = plt.gca()

plt.plot(sol.y[0], sol.y[1])
ax.set_aspect('equal', 'box')
```
